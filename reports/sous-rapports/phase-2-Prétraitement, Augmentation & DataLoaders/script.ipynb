# Chapitre 2 : Prétraitement, Augmentation et Flux de Données (*DataLoaders*)

## 2.1 Introduction à la Phase de Préparation
Une fois l'exploration initiale achevée, la seconde phase du projet consiste à transformer le corpus brut en un flux de données optimisé pour l'apprentissage profond (*Deep Learning*). Cette étape ne se limite pas à un simple nettoyage ; elle définit la stratégie de généralisation du modèle face à la diversité clinique et assure l'étanchéité statistique entre les phases d'entraînement et d'évaluation.

## 2.2 Partitionnement des Données et Étanchéité Statistique
Le premier défi technique relevé a été la sécurisation du partitionnement des données. Contrairement à une approche aléatoire standard, nous avons implémenté une séparation par groupe afin de neutraliser le risque de fuite de données (*Data Leakage*).

* **Méthodologie :** Utilisation de l'algorithme `GroupShuffleSplit` (Scikit-Learn).
* **Critère de partitionnement :** Variable `lesion_id`.
* **Répartition :** 80 % Entraînement / 10 % Validation / 10 % Test.
* **Résultat de l'Audit :** L'analyse d'intersection a confirmé une étanchéité totale, garantissant qu'aucune lésion utilisée pour l'entraînement n'influence l'évaluation finale.

> **[Résultats de l'output : Cellule 2 & Cellule 4 — "Audit de Sécurité : Succès Total"]**

## 2.3 Ingénierie des Données Tabulaires et Persistance
Pour le modèle de forêt aléatoire (*Random Forest*), les métadonnées cliniques ont subi un traitement spécifique visant à assurer la stabilité numérique et la compatibilité avec un déploiement réel.

* **Imputation :** Les lacunes sur la variable « Âge » ont été comblées par la médiane du jeu d'entraînement.
* **Normalisation et Encodage :** L'application de `StandardScaler` et `OneHotEncoder` a permis de projeter les variables cliniques dans un espace mathématique exploitable.
* **Persistance (MLOps) :** Les objets de transformation ont été sérialisés au format `.pkl` (joblib). Cette approche garantit que l'application finale traitera les futurs patients avec les mêmes paramètres statistiques que ceux du modèle original.

> **[Documents générés : Cellule 10 — `models/encoder.pkl`, `models/scaler.pkl`]**

## 2.4 Architecture du Pipeline de Vision : Augmentation « à la volée »
Le traitement des images repose sur une architecture de transformation dynamique. Pour maximiser la robustesse du réseau de neurones convolutif (CNN), nous avons implémenté une augmentation de données en temps réel.

* **Algorithme de Redimensionnement :** Utilisation de la méthode `Resize(232)` suivie d'un `RandomCrop(224)`. Ce choix stratégique préserve le ratio d'aspect des lésions cutanées, évitant les distorsions morphologiques qui pourraient fausser le diagnostic.
* **Variabilité Géométrique et Colorimétrique :** L'introduction de rotations aléatoires, de symétries axiales et de variations de luminosité (`ColorJitter`) permet de simuler la diversité des conditions d'examen dermoscopique.

> **[Code source : Cellule 12 — Configuration `train_transforms`]**

## 2.5 Stratégie d'Équilibrage Dynamique par Échantillonnage Pondéré
Face au déséquilibre massif entre les classes bénignes et malignes, nous avons opté pour une technique d'échantillonnage préférentiel plutôt qu'une simple modification de la fonction de perte.

* **Mécanisme :** Implémentation du `WeightedRandomSampler` (PyTorch).
* **Fonctionnement :** Chaque échantillon se voit attribuer un poids inversement proportionnel à la fréquence de sa classe. Ce mécanisme force le système à présenter, statistiquement, autant d'exemples de mélanomes que de nævus lors de chaque cycle d'apprentissage (*Epoch*).
* **Synergie :** Combiné à l'augmentation de données, ce procédé permet au modèle de voir les images rares sous de multiples angles, compensant ainsi le manque de données initiales sans risquer un surapprentissage brutal.

## 2.6 Validation Visuelle des Flux (*DataLoaders*)
La dernière étape a consisté à valider l'intégrité du moteur d'alimentation. L'inspection des lots de données (*Batches*) a confirmé l'efficacité des transformations et la réussite de l'équilibrage des classes.

Ce pipeline de données constitue une infrastructure robuste. En isolant les lésions, en automatisant l'augmentation de données et en garantissant la persistance des paramètres cliniques, nous avons créé un environnement propice à un entraînement rigoureux.
